```python
%matplotlib inline

import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
```

____
Distribuição teórica escolhida: Poisson
#### X~Poisson(y)
OBS.: y é igual a E(X) e Var(X) = 5

```python
x = stats.poisson.rvs(5, loc = 0, size = 1000000)
E = np.mean(x)
Var = np.var(x)

print("Média: {0}".format(E))
print("Variância: {0}".format(Var))
```

```python
p = np.random.poisson(5, 1000000)

count, bins, ignored = plt.hist(p, 14, normed=True)
plt.show()
```

```python
ps = pd.Series(data = sorted(p))
plt.plot(ps, stats.poisson.pmf(5, ps))
```

```python
n = 2

rc = np.random.choice(x,n)
print(rc)
```

```python
def media(rc):
    soma= 0
    for i in range(len(rc)):
        soma += rc[i]
    media = soma/len(rc)
    return media

f = media(rc)
print(f)
```

```python
def func_m(x,n):
    lista = []
    for i in range(10000):
        rc = np.random.choice(x,n)
        f = media(rc)
        lista.append(f)
    return lista

func = func_m(x,n)
print(func)
```

```python
h = pd.Series(data=sorted(func))
```

```python
h.describe()
```

```python
faixa = np.arange(0,13,0.5)
```

```python
plot = h.plot.hist(bins=faixa, normed=True)
```

```python
def X(x,n3):
    X1 = []
    X9 = []
    for i in range(10000):
        rc = np.random.choice(x,n3)
        X1.append(rc[0])
        X9.append(rc[8])
    return (X1, X9)

X19 = X(x,n3)

X1 = X19[0]
X9 = X19[1]
```

```python
h4 = pd.Series(data=sorted(X1))
faixa4 = np.arange(0,(h4.max()+1),1)
plot = h4.plot.hist(bins=faixa4, normed=True)
plt.plot(h4, stats.poisson.pmf(5, h4), color="r")
```

```python
h5 = pd.Series(data=sorted(X9))
faixa5 = np.arange(0,(h5.max()+1),1)
plot = h5.plot.hist(bins=faixa5, normed=True)
plt.plot(h5, stats.poisson.pmf(5, h5), color="r")
```
